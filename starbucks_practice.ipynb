## Create a Combined Vitamin Column. Find the Drinks Contain highest or lowest Vitamin. 

```python
#df[vit_A_perc_DV] = df[vit_A_perc_DV].str.replace('%','').astype('float')/100.0
df['vit_A_perc_DV'] = df['vit_A_perc_DV'].str.rstrip('%').astype('float')/100.0
df['vit_C_perc_DV'] = df['vit_C_perc_DV'].str.rstrip('%').astype('float')/100.0
```

```python
df['vit_total'] = df['vit_A_perc_DV'] + df['vit_C_perc_DV']
df
```

```python
beverage_vit = df.pivot_table(index = 'beverage', values = 'vit_total')
beverage_vit.sort_values('vit_total', ascending = False, inplace = True)
beverage_vit
```
